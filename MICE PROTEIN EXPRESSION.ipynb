```python
import pandas as pd
data = pd.read_csv('Data_Cortex_Nuclear.csv')
data.drop(['MouseID',],axis=1, inplace=True)
pd.DataFrame(data[:10])
```

```python
data = data.drop(['Treatment', 'Behavior', 'class'],axis=1)
data.fillna(value=0)
```

```python

data.isnull().sum()
```

```python
data.corr()
```

```python
import numpy as np
columns = data.columns
X_data = data[columns[:-1]]
y_data = data[columns[-1]]
X_data.replace('',np.NaN,inplace=True)
from sklearn.preprocessing import Imputer
imputer = Imputer()
imputer.fit(X_data)
#THIS IS A MUST OR THE LABELS WITH BE NUMERICAL AND THE PROTEIN EXPRESSION WILL BE UNKNOWN TO MOST OF US
X_data = pd.DataFrame(columns=X_data.columns,data=imputer.transform(X_data))
pd.DataFrame(X_data[:10])
```

```python
X_data.isnull().sum()
```

```python
X_data.corr()
```

```python
from sklearn.model_selection  import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_data,y_data,test_size=.3)
print(X_train.shape)
print(X_train[:5])
print(y_train.shape)
print(y_train[:5])
```

```python
print(X_test[:5])
print(y_test)
```

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(solver='adam', validation_fraction=.3,alpha=.000055, hidden_layer_sizes=(25,25,8), random_state=1,tol=.000000099,verbose=True)

clf.fit(X_train, y_train)
y_train.shape
```

```python
clf.score(X_test,y_test)
```

```python
print(clf.predict(X_test[:10]))
print(y_test[:10])
```

```python
pd.DataFrame(clf.predict_proba(X_test[:10]))
```
